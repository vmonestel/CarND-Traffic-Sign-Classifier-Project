````python

import cv2
def image_scale(img):
    scale_factor = random.randint(1, 2)
    if (scale_factor == 1):
        return img
    result_img = cv2.resize(img, None, fx=scale_factor, fy=scale_factor)
    center_x, center_y = int(result_img.shape[0]/2), int(result_img.shape[1]/2)
    result_img = result_img[center_x - 16:center_x + 16, center_y - 16:center_y + 16]
    return result_img

def image_rotate(img):
    rows, cols, channels = img.shape
    degrees = random.randint(-5, 5)
    M = cv2.getRotationMatrix2D((cols/2, rows/2), degrees, 1)
    result_img = cv2.warpAffine(img,M,(cols,rows))
    return result_img

def image_translate(img):
    rows, cols, channels = img.shape
    col_pixels = random.randint(-2, 2)
    row_pixels = random.randint(-2, 2)
    M = np.float32([[1, 0, col_pixels], [0, 1, row_pixels]])
    result_img = cv2.warpAffine(img, M, (cols,rows))
    return result_img

def image_brightness_change(img):
    #a = np.double(img)
    change = random.randint(-8, 8)
    a = img - change
    result_img = np.uint8(a)
    return result_img

index = random.randint(0, len(X_train))
image = X_train[index].squeeze()
plt.figure(figsize=(4,4))
plt.imshow(image)
print(image.shape)

#result = (image_rotate(image)).squeeze()
#print(result.shape)
#plt.figure(figsize=(4,4))
#plt.imshow(result)
#print(result.shape)

#result = (image_scale(image)).squeeze()
#plt.figure(figsize=(4,4))
#plt.imshow(result)
#print(result.shape)

#print("B change")
result = (image_brightness_change(image)).squeeze()
plt.figure(figsize=(4,4))
plt.imshow(result)
print(result.shape)

result = (image_scale(image_translate(image_rotate(image)))).squeeze()
print(result.shape)
plt.figure(figsize=(4,4))
plt.imshow(result)
print(result.shape)
````

### Model Architecture

````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten

# Define EPOCHS
EPOCHS = 50
# Define BATCH_SIZE (using suggested value)
BATCH_SIZE = 128

# Arguments used for tf.truncated_normal
mu = 0
sigma = 0.1
````

````python
# Define neural network design
# LeNet provides the following layers:
# Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6. Relu Activation
# Layer 2: Convolutional. Output = 10x10x16. Relu Activation
# Layer 3: Fully Connected. Input = 400. Output = 120.
# Layer 4: Fully Connected. Input = 120. Output = 84.
# Layer 5: Fully Connected. Input = 84. Output = 10.
def LeNet(x, keep_prob):
    # Randomly defines variables for the weights and biases for each layer
    weights = {
    'layer_1_weights': tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean = mu, stddev = sigma)),
    'layer_2_weights': tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma)),
    'layer_3_weights': tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma)),
    'layer_4_weights': tf.Variable(tf.truncated_normal(shape=(120, 84), mean = mu, stddev = sigma)),
    'out_layer_weights': tf.Variable(tf.truncated_normal(shape=(84, n_classes), mean = mu, stddev = sigma))}

    biases = {
    'layer_1_biases': tf.Variable(tf.random_normal([6])),
    'layer_2_biases': tf.Variable(tf.random_normal([16])),
    'layer_3_biases': tf.Variable(tf.random_normal([120])),
    'layer_4_biases': tf.Variable(tf.random_normal([84])),
    'out_layer': tf.Variable(tf.random_normal([n_classes]))}
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    conv_1 = tf.nn.conv2d(x, weights['layer_1_weights'], strides=[1, 1, 1, 1], padding='VALID')
    conv_1 = tf.nn.bias_add(conv_1, biases['layer_1_biases'])

    # Activation.
    conv_1 = tf.nn.relu(conv_1)

    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv_1 = tf.nn.max_pool(conv_1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # Layer 2: Convolutional. Output = 10x10x16.
    conv_2 = tf.nn.conv2d(conv_1, weights['layer_2_weights'], strides=[1, 1, 1, 1], padding='VALID')
    conv_2 = tf.nn.bias_add(conv_2, biases['layer_2_biases'])
    
    # Activation.
    conv_2 = tf.nn.relu(conv_2)

    # Pooling. Input = 10x10x16. Output = 5x5x16.
    conv_2 = tf.nn.max_pool(conv_2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # Flatten. Input = 5x5x16. Output = 400.
    flat   = flatten(conv_2)
    
    # Layer 3: Fully Connected. Input = 400. Output = 120.
    fc1 = tf.add(tf.matmul(flat, weights['layer_3_weights']), biases['layer_3_biases'])
    
    # Activation.
    fc1 = tf.nn.relu(fc1)
    fc1 = tf.nn.dropout(fc1, keep_prob)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2 = tf.add(tf.matmul(fc1, weights['layer_4_weights']), biases['layer_4_biases'])
    
    # Activation.
    fc2 = tf.nn.relu(fc2)
    fc2 = tf.nn.dropout(fc2, keep_prob)

    # Layer 5: Fully Connected. Input = 84. Output = 10.
    logits = tf.add(tf.matmul(fc2, weights['out_layer_weights']), biases['out_layer'])
    
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
# Define tensor flow variables
keep_prob = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
````

````python
# Define learning rate
rate = 0.0009
# Use one hot enconding
one_hot_y = tf.one_hot(y, 43)
# Get the logits from Lenet model
logits = LeNet(x, keep_prob)
# Optimize the loss using Adam optimizer
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()
````

````python
# Function that calculates the accuracy of a data set
def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
# Train the model
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.75})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
````
